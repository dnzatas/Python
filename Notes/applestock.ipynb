```python
import yfinance as yf
import json
```

```python
apple = yf.Ticker('AAPL')
print(apple)
```

```python
print(apple.info)
```

<p>period: data period to download (Either Use period parameter or use start and end) Valid periods are: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max</p>
<p>interval: data interval (intraday data cannot extend last 60 days) Valid intervals are: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo</p>
<p>start: If not using period - Download start date string (YYYY-MM-DD) or datetime.</p>
<p>end: If not using period - Download end date string (YYYY-MM-DD) or datetime.</p>
<p>prepost: Include Pre and Post market data in results? (Default is False)</p>
<p>auto_adjust: Adjust all OHLC automatically? (Default is True)</p>
<p>actions: Download stock dividends and stock splits events? (Default is True)</p>

```python
apple_history = apple.history(period='max')
apple_history
```
