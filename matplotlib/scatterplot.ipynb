```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn-v0_8')
```

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

plt.scatter(x, y)
```

```python
plt.scatter(x, y, s=200, c='green') # s is PLOT SIZE. c is PLOT color
```

```python
plt.scatter(x, y, s=100, c='red', marker='x')
```

```python
# edgecolors: color of circle around
# linewidths: the thickness of the line around the circle
# aplha: soften edge color

plt.scatter(x,y, s=200, c="green", edgecolors="black", linewidths=1, alpha=0.75)
```

```python
colors = [7,5,3,7,5,9,1,5,6,2,6,8,2,9,1,5,2,8,2,9] # different numbers to different possibilites from 1 to 10

plt.scatter(x,y, s=100, c=colors, edgecolors='black', linewidths=1, alpha=0.75)
```

```python
# with cmap ans colors we get different shades of Green. The lighter close to 0 and dark ones close to 10

plt.scatter(x,y, s=100, c=colors, cmap="Greens", edgecolors="black", linewidths=1, alpha=0.75)
```

```python
plt.scatter(x,y, s=100, c=colors, cmap="Greens", edgecolors="black", linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539] # sizes of plots

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', edgecolors='black', linewidths=1, alpha=0.75)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv")
df.head()

# df.to_csv("dosya_adi.csv") # you can save the data that you have fetch from github as csv
```

```python
print(df.shape)
df.isnull().any()
```

```python
x = df["view_count"]
y = df["likes"]
colors = df["ratio"]

plt.scatter(x, y, c=colors, cmap="summer", edgecolors="black", linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")

plt.xscale("log")
plt.yscale("log")

plt.title("Youtube Trend Videos")
plt.xlabel("View Count")
plt.ylabel("Likes Count")
```
